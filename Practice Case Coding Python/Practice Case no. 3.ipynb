```python
import json
import csv

f = '''
[
   {
       "user": {
           "id": 1,
           "name": "agus",
           "email": "agus@example.com"
       },
       "txs": [
           {
               "id": 1,
               "tx_date": "2020-01-01",
               "tx_amount": 10,
               "tx_type": "buy"
           },
           {
               "id": 3,
               "tx_date": "2020-02-01",
               "tx_amount": 10,
               "tx_type": "buy"
           },
           {
               "id": 5,
               "tx_date": "2020-03-01",
               "tx_amount": 20,
               "tx_type": "sell"
           }
       ]
   },
   {
       "user": {
           "id": 2,
           "name": "asep",
           "email": "asep@example.com"
       },
       "txs": [
           {
               "id": 2,
               "tx_date": "2020-01-01",
               "tx_amount": 10,
               "tx_type": "sell"
           },
           {
               "id": 4,
               "tx_date": "2020-02-01",
               "tx_amount": 10,
               "tx_type": "sell"
           },
           {
               "id": 6,
               "tx_date": "2020-03-01",
               "tx_amount": 20,
               "tx_type": "buy"
           }
       ]
   }
]
'''

raw_data = json.loads(f)

with open('CSV_1.csv', mode='w', newline='') as csv_1_file:
    writer = csv.writer(csv_1_file, delimiter=',', quotechar='|', quoting=csv.QUOTE_MINIMAL)

    writer.writerow(['user_id', 'tx_id'])
    for id in range(len(raw_data)):
        for txid in range(len(raw_data[id]['txs'])):
            new_line = [raw_data[id]['user']['id'], raw_data[id]['txs'][txid]['id']]
            writer.writerow(new_line)

with open('CSV_2.csv', mode='w', newline='') as csv_2_file:
    fieldnames = raw_data[0]['user'].keys()
    
    writer = csv.DictWriter(csv_2_file, fieldnames=fieldnames)
    writer.writeheader()
    for id in range(len(raw_data)):
        writer.writerow(raw_data[id]['user'])

txs_list = []
for id in range(len(raw_data)):
    txs_list += raw_data[id]['txs']
txs_list_sorted = sorted(txs_list, key = lambda i: i['id'])

with open('CSV_3.csv', mode='w', newline='') as csv_3_file:
    fieldnames = raw_data[0]['txs'][0].keys()
    
    writer = csv.DictWriter(csv_3_file, fieldnames=fieldnames)
    writer.writeheader()
    for txid in range(len(txs_list_sorted)):
        writer.writerow(txs_list_sorted[txid])

```
