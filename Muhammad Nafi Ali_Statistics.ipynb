# Import library dan csv

```python
import pandas as pd
from sklearn import preprocessing

raw_data = pd.read_csv('Seasons_Stats.csv')

data = raw_data.copy()
```

# Preprocessing

```python
data = data[data['Year'] == 2017]
data = data.drop_duplicates(subset = 'Player', keep = 'last', ignore_index = True)
data = data.drop(['blanl', 'blank2'], axis = 1)
```

# Mencari pemain paling muda

```python
team_young = data[data.groupby(by = 'Tm')['Age'].transform(min) == data['Age']].sort_values(by = 'Tm')
team_young
```

# Mencari pemain paling tua

```python
team_old = data[data.groupby(by = 'Tm')['Age'].transform(max) == data['Age']].sort_values(by = 'Tm')
team_old
```

# Mencari pemain dengan  MP paling tinggi

```python
mp_max = data[data.groupby(by = 'Pos')['MP'].transform(max) == data['MP']].sort_values(by = 'Pos')
mp_max
```

# Mencari pemain dengan nilai fitur paling tinggi

```python
team_average = data.groupby(by = 'Tm').mean()
team_average = team_average.reset_index()
team_average.loc[team_average[team_average['TRB%'] == team_average['TRB%'].dropna().max()].index].loc[:, ['Tm', 'TRB%']]
```

```python
team_average.loc[team_average[team_average['AST%'] == team_average['AST%'].dropna().max()].index].loc[:, ['Tm', 'AST%']]
```

```python
team_average.loc[team_average[team_average['STL%'] == team_average['STL%'].dropna().max()].index].loc[:, ['Tm', 'STL%']]
```

```python
team_average.loc[team_average[team_average['BLK%'] == team_average['BLK%'].dropna().max()].index].loc[:, ['Tm', 'BLK%']]
```

# Mencari pemain terbaik

```python
data_normalize = data.loc[:, ['PTS', 'TRB', 'AST', 'STL', 'BLK']]
data_normalize = data_normalize.T.div(data.loc[:,'MP']).T
data_normalize = preprocessing.normalize(data.loc[:, ['PTS', 'TRB', 'AST', 'STL', 'BLK']], axis = 0)
data_normalize = pd.DataFrame(data_normalize, columns = ['PTS', 'TRB', 'AST', 'STL', 'BLK'])
data_normalize['Total'] = data_normalize.sum(axis = 1)
data.loc[data_normalize[data_normalize['Total'] == data_normalize['Total'].max()].index]
```

# Mencari tim terbaik

```python
data_normalize_team = data.copy().loc[:,'Tm']
data_normalize_team = pd.concat([data_normalize_team, data_normalize], axis = 1)
data_normalize_team = data_normalize_team.groupby(by = 'Tm').mean()
data_normalize_team = data_normalize_team.reset_index()
index_max = data_normalize_team[data_normalize_team['Total'] == data_normalize_team['Total'].dropna().max()].index
data_normalize_team.loc[index_max].loc[:, ['Tm', 'Total']]
```
